```python
x = np.array([[1,2],
    [3,4],
    [5,6],
     [7,8]
    ])
y = np.array([1,2,3,4])
```

```python
m1 = 0
m2 = 0
c = 0
alpha = 0.01
iter = 1000
```

```python
n = len(x)
```

```python
for i in range(iter):
    dm1 = 0
    dm2 = 0
    dc = 0
    for i in range(n):
        x1 = x[i][0]
        x2 = x[i][1]
        y_pred = m1*x1 + m2*x2+c
        error = y_pred -y[i]

        dm1 +=error*x1
        dm2 +=error*x2
        dc += error
    dm1 = dm1/n
    dm2 = dm2/n
    dc = dc/n

    m1 = m1 - alpha*dm1
    m2 = m2-alpha*dm2
    c = c-alpha*dc 
```
