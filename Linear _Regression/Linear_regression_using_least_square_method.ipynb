```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4, 5], dtype=float)
Y = np.array([2, 4, 6, 8, 10,], dtype=float)
```

```python
sum_x = np.sum(X)
sum_y = np.sum(Y)
sum_x_sq = np.sum(X**2)
sum_X_mul_Y = np.sum(X*Y)
```

```python
n = len(X)
```

```python
m = (n*sum_X_mul_Y -((sum_x)*(sum_y)))/(n*(sum_x_sq)-((sum_x)**2))
c = (sum_y- (m*(sum_x)))/n
```

```python
y_pred = m*X+c
```

```python
print(m)
```

```python
print(c)
```

```python
plt.scatter(X,Y,label = 'Actual data') 
plt.plot(X,y_pred,label = 'Best fit line')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
